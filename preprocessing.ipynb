```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 200)
```

```python
housing = pd.read_csv('data/Ames_HousePrice.csv', index_col=0)
housing.shape
```

```python
housing.head()
```

```python
housing.dtypes
```

```python
housing.isnull().sum(axis=0)
```

```python
price = housing['SalePrice']
features = housing.drop(["PID", "SalePrice"], axis=1)
```

```python
num_features = features.select_dtypes(include=['int64','float64'])
num_features.shape
```

```python
num_features.isna().sum()
```

```python
# LotFrontage: feet of street connected to property - probably not 0, if there is a yard at all
# MasVnrArea: area of veneer - could be 0? look at veneer type
# bsmt sq feet, bathrooms - look at other basement variables
# garageyrbuilt, garagecars, garagearea - look at if there is a garage
```

```python
num_features.LotFrontage.describe() # minimum 21, max 313 (no 0s)
```

```python
# for those with missing lot frontage values, lot area ranges from 1533-164669 sq ft
num_features[num_features['LotFrontage'].isna()].LotArea.describe()
```

```python
# determine relationship between lot area and lot frontage to impute missing values
lot = num_features[num_features['LotFrontage'].notna()]
x = lot['LotFrontage']
y = lot['LotArea']
m, b = np.polyfit(x, y, 1)
plt.scatter(x, y)
plt.plot(x, m*x+b, color='red')
```

```python
# impute missing values using the mean relationship between lot area and lot frontage 
m # 134
num_features['LotFrontage'].fillna(num_features['LotArea']/m, inplace = True)
```

```python
num_features.LotFrontage.describe()
```

```python
# some of the imputed values appear to be outliers
plt.scatter(num_features['LotFrontage'], num_features['LotArea'])
```

```python
# for missing values masonry veneer area, there is no masonry veneer
housing[housing['MasVnrArea'].isna()].MasVnrType
```

```python
# impute with 0s
num_features['MasVnrArea'].fillna(0, inplace = True)
```

```python
# for missing values of Bsmt variables, there is no basement
housing[housing['BsmtFullBath'].isna()].filter(regex='Bsmt')
```

```python
# impute with 0s
num_features.fillna({'BsmtFinSF1': 0, 'BsmtFinSF2': 0, 'BsmtUnfSF': 0, 'TotalBsmtSF': 0,
                    'BsmtFullBath': 0, 'BsmtHalfBath': 0},inplace=True)
```

```python
# view missing garagecars/garagearea
housing[housing['GarageArea'].isna()].filter(regex='Garage')
```

```python
# impute with mean values
num_features.fillna({'GarageArea': num_features.GarageArea.mean(), 
                     'GarageCars': num_features.GarageCars.mean()},inplace=True)
```

```python
# for missing garage year built, most don't have a garage
housing[housing['GarageYrBlt'].isna()].filter(regex='Garage')
```

```python
# impute garage year built with overall house year built
num_features.fillna({'GarageYrBlt': num_features.YearBuilt},inplace=True)
```

```python
num_features.isna().sum()
```

```python
# convert MsSubClass to categorical variable
num_features['MSSubClass'].value_counts()

num_features['MSSubClass'] = num_features['MSSubClass'].replace({20: '1S_1946newer', 
                                                                 30: '1S_1945older', 
                                                                 40: '1S_finishedattic',
                                                                 45: '1_2S_unfinished',
                                                                 50: '1_2S_finished',
                                                                 60: '2S_1946newer',
                                                                 70: '2S_1945older',
                                                                 75: '2halfS',
                                                                 80: 'splitlevel',
                                                                 85: 'splitfoyer', 
                                                                 90: 'duplex',
                                                                 120: '1S_PUD',
                                                                 150: '1halfS_PUD',
                                                                 160: '2S_PUD',
                                                                 180: 'Multi_PUD',
                                                                 190: '2Fam_conversion'
                                                                })
num_features['MSSubClass'].value_counts()
```

```python
# explore missing values for categorical variables
obj_features = features.select_dtypes(include=['object'])
obj_features.shape
```

```python
obj_features.isna().sum()
```

```python
# according to the data dictionary, NA values represent none of that attribute
# impute all with None
obj_features.fillna('None', inplace = True)
```

```python
housing_nona = pd.concat([num_features, obj_features, price], axis = 1)
```

```python
# addressing outliers
```

```python
# interquartile range

# calculate interquartile range and the outlier cutoff
q25, q75 = np.percentile(housing_nona['SalePrice'], 25), np.percentile(housing_nona['SalePrice'], 75)
iqr = q75 - q25
cut_off = iqr * 1.5
lower, upper = q25 - cut_off, q75 + cut_off
print('Percentiles: 25th=%.3f, 75th=%.3f, IQR=%.3f' % (q25, q75, iqr))

# identify outliers
outliers = [x for x in housing_nona['SalePrice'] if x < lower or x > upper]
print('Identified outliers: %d' % len(outliers))
```

```python
housing_new = housing_nona[(housing_nona['SalePrice'] < upper) & (housing_nona['SalePrice'] > lower)]
```

```python
housing_new.shape
```

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].hist(housing_nona['SalePrice'])

#axs[0, 0].set_title("main")
axs[1, 0].boxplot(housing_nona['SalePrice'], vert=False)
#axs[1, 0].set_title("shares x with main")
#axs[1, 0].sharex(axs[0, 0])

axs[0, 1].hist(housing_new['SalePrice'])
#axs[0, 1].set_title("unrelated")
axs[1, 1].boxplot(housing_new['SalePrice'], vert=False)
#axs[1, 1].set_title("also unrelated")

#fig.tight_layout()
```

```python
# export as csv
housing_nona.to_csv('data/houseprice_clean.csv', index=False)
housing_new.to_csv('data/houseprice_trimmed.csv', index=False)
```

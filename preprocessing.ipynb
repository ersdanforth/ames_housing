```python
# LotFrontage: feet of street connected to property - probably not 0, if there is a yard at all
# MasVnrArea: area of veneer - could be 0? look at veneer type
# bsmt sq feet - look at other basement variables
# bsmt full/half bath - could be 0
# garageyrbuilt, garagecars, garagearea - look at if there is a garage
```

```python
num_features.LotFrontage.describe() # minimum 21, max 313
```

```python
# for those with missing lot frontage values, lot area ranges from 1533-164669 sq ft
num_features[num_features['LotFrontage'].isna()].LotArea.describe()
```

```python
# determine relationship between lot area and lot frontage to impute missing values

lot = num_features[num_features['LotFrontage'].notna()]
x = lot['LotFrontage']
y = lot['LotArea']
m, b = np.polyfit(x, y, 1)

plt.scatter(x, y)
plt.plot(x, m*x+b, color='red')
```

```python
m
```

```python
# impute missing values using the mean relationship between lot area and lot frontage 
```

```python
#num_features[num_features['LotFrontage'].isna()].LotFrontage 
num_features['LotFrontage'].fillna(num_features['LotArea']/134.18, inplace = True)
```
